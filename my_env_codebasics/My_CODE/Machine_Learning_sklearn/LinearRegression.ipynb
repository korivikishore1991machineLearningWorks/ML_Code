```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\koriv\\Desktop\\MachineLearning_DataScience\\Hands_On_Machine_Learning\\canada_per_capita_income.csv')
df
```

```python
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.replace('\\$','').str.replace('\\(','').str.replace('\\)','')
df.columns
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df.year,df.per_capita_income_US,color='red',marker='+')
```

```python
new_df = df.drop('per_capita_income_US',axis='columns')
new_df
```

```python
per_capita_income_US = df.per_capita_income_US
per_capita_income_US
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,per_capita_income_US)
```

```python
reg.predict([[2020]])
```

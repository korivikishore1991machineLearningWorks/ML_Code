```python
# Ordinal Encoding for person_home_ownership and cb_person_default_on_file for Catogorical Ordinal Variables (lable encoder is for Target)
## cols_to_be_OE
col_catO = ['person_home_ownership', 'cb_person_default_on_file']

## Handeling missing value via droping the null rows for cb_person_default_on_file column
X = X[X[col_catO[1]].notnull()]

pp_O1 = Pipeline([
    ('person_home_ownership_Imputer', SimpleImputer(strategy='constant', add_indicator=False, fill_value='OTHER')),
])
pp_02 = Pipeline([
    ('oe_cat', OrdinalEncoder(categories=[person_home_ownership_unique, cb_person_default_on_file_unique]))
])
    
ct_catO = ColumnTransformer([
    ('pp_O1', pp_O1, [col_catO[1]]),
    ('pp_02', pp_02, col_catO)
], remainder='passthrough')
    
xt = ct_catO.fit_transform(X)
xt
```

```python
# One hot Encoding for loan_intent and loan_grade for Catogorical Nominal Variables

## Define a DataPipe line for filling missing values and then encoding using One hot encoding
pp_catN = Pipeline(steps=[
    ('imp_catN', SimpleImputer(strategy='constant', add_indicator=False, fill_value='missing')),
    ('ohe_catN', OneHotEncoder(sparse=False, handle_unknown='ignore'))
])

## cols_to_be_OHE
col_catOHE = ['loan_intent', 'loan_grade']

## Define a column Transformer
ct_catN = ColumnTransformer(transformers=[
    ('pp_catN', pp_catN, col_catOHE)
], remainder='passthrough')

## Transform the X train
ct_catN.fit(X)
xt = ct_catN.transform(X)
display(xt)
#display(xt.get_feature_names())
display(xt[0])
len(xt[0])


#Visualize the transformed features
#xt.transformers[0][1][1].get_feature_names()
#ct_catN.transformers_[0]
enc_cat_features = ct_catN.named_transformers_['pp_catN']['ohe_catN'].get_feature_names()
numeric_features = X.select_dtypes(np.number).columns
numeric_features
labels = np.concatenate([enc_cat_features, numeric_features,])
transformed_df_columns = pd.DataFrame(xt, columns=labels)
transformed_df_columns
```
